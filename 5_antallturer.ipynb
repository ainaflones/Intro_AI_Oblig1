```python
#Importerer bibliotekene vi skal bruke
import pandas as panda
from matplotlib import pyplot as plt

#Laster data fra fil og viser de 5 første radene
dataframe = panda.read_csv("Ruter_data.csv", delimiter=";")
dataframe.head()

#Representerer ønsket data i et barchart
dataframe['Kommune'].value_counts().sort_values(ascending=False).plot.bar(width=0.9)
plt.title("Antall turer fordelt på kommuner fra juni til august, sortert")
plt.xlabel("Kommuner")
plt.ylabel("Antall turer")
```
