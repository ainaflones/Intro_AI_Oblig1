```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
dataframe = pd.read_csv("/Users/ainaflones/Downloads/Ruter_data.csv", delimiter=";")

#Choosing rows where the arrival time is later than expected arrival, AKA where the bus is late.
dataframe2 = dataframe[dataframe['Tidspunkt_Faktisk_Ankomst_Holdeplass_Fra'] > dataframe['Tidspunkt_Planlagt_Ankomst_Holdeplass_Fra']]

#Using a bar graph to visualise how many times during the day the bus is late. Sorting this by amount of times late in a day. 
dataframe2['Dato'].value_counts().sort_values(ascending=False).plot.bar(width=0.9)
plt.ylabel("Antall")
plt.xlabel("Dato")
plt.show
```
