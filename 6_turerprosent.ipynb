```python
#Importerer bibliotekene vi skal bruke
import pandas as panda
from matplotlib import pyplot as plt

#Laster data fra fil og viser de 5 første radene
dataframe = panda.read_csv("Ruter_data.csv", delimiter=";")
dataframe.head()
```

```python
dataframe['Område'].value_counts().plot.pie(startangle=90, autopct='%1.1f%%')
plt.title('Turer fordelt på områdene i prosent')
plt.ylabel('')
```
